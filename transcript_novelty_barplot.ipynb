```r
library(tidyverse)
```

```r
read_annot = read.delim("data/all_talon_abundance_filtered.tsv")
```

```r
read_annot$transcript_novelty <- factor(read_annot$transcript_novelty, 
                                        levels = c("Known", "ISM", "NIC",
                                                   "NNC","Antisense","Intergenic","Genomic"))

ggplot(data = read_annot) + 
    geom_bar(mapping = aes(x = transcript_novelty, fill = transcript_novelty)) + 
    theme(axis.text.x = element_text(angle = 45, hjust = 1)) + 
    scale_fill_manual(values = c("#009E73","#0072B2","#D55E00","#E69F00","#000000","#CC79A7","#F0E442")) +
    ylab("# of distinct transcript models") + theme_minimal()
```
